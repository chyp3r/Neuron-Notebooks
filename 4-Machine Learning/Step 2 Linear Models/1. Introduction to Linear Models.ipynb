# Introduction to Linear Models

Linear models are the simplest form of machine learning models, designed to establish a linear relationship between input features (predictors) and the target variable. They form the foundation of many complex algorithms and are widely used in both regression and classification problems.

## 1. What is a Linear Model?

A linear model assumes a linear relationship between the input variables (X) and the output variable (Y). The equation for a linear model is:

**y = w₀ + w₁x₁ + w₂x₂ + ... + wₙxₙ**

Where:
- **y** is the predicted target value.
- **w₀** is the intercept (bias term).
- **w₁, w₂, ... wₙ** are the weights (coefficients) for each input feature.
- **x₁, x₂, ... xₙ** are the input features.

Linear models aim to find the best weights (coefficients) that minimize the error between the predicted and actual values.


## 2. Linear Models in Machine Learning
Linear models are used in both **regression** and **classification** tasks:

- **Linear Regression**: Used when the target variable is continuous (e.g., predicting house prices).
- **Logistic Regression**: Used for binary classification problems (e.g., predicting whether an email is spam or not).

Linear models are popular due to their simplicity and interpretability. However, they may not always capture complex relationships in the data.


## 3. When to Use Linear Models
Linear models are most appropriate when:
- You believe the relationship between input variables and the target is approximately linear.
- The dataset is not too large, and the model needs to be simple and interpretable.
- The data has no strong non-linear interactions between variables.

However, they are less suitable for datasets with complex, non-linear patterns.


## 4. Example: Linear Relationship

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

np.random.seed(42)
X = 2 * np.random.rand(100,1)
y = 4+3*X +np.random.rand(100,1)

model = LinearRegression()
model.fit(X,y)

y_pred = model.predict(X)

plt.scatter(X,y,color= "blue")
plt.plot(X,y_pred,color="red")
plt.show()

print(f"Intercept (w₀): {model.intercept_}")
print(f"Coefficient (w₁): {model.coef_}")
```
