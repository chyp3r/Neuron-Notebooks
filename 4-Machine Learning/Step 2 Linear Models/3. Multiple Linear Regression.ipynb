# Introduction to Multiple Linear Regression

Multiple Linear Regression is an extension of Simple Linear Regression where we predict a dependent variable (target) based on multiple independent variables (predictors). The relationship is modeled as a linear equation, allowing us to explore the combined effect of multiple features on the target variable.

## 1. What is Multiple Linear Regression?

Multiple Linear Regression models the relationship between two or more predictors (independent variables) and a target (dependent variable). The formula is:

**y = w₀ + w₁x₁ + w₂x₂ + ... + wₙxₙ + ε**

Where:
- **y**: the dependent variable (target)
- **x₁, x₂, ..., xₙ**: the independent variables (predictors)
- **w₀**: the intercept (bias term)
- **w₁, w₂, ..., wₙ**: the coefficients (weights) associated with each independent variable
- **ε**: the error term (residual)

---

## 2. Assumptions of Multiple Linear Regression

1. **Linearity**: The relationship between the independent and dependent variables is linear.
2. **Independence**: Observations should be independent.
3. **Homoscedasticity**: The variance of the residuals is constant.
4. **No multicollinearity**: The independent variables are not highly correlated with each other.
5. **Normality**: The residuals are normally distributed.

---

## 3. Multiple Linear Regression Formula

The general formula for Multiple Linear Regression is:

**y = w₀ + w₁x₁ + w₂x₂ + ... + wₙxₙ**

Our goal is to find the coefficients (**w₀, w₁, w₂, ..., wₙ**) that minimize the sum of squared residuals.

---

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
X,y = make_regression(n_samples=1000,n_features=3)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.3,random_state=42)

model = LinearRegression()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test,y_pred)
plt.title("Predicted vs Actual")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.plot([min(y_pred), max(y_pred)], [min(y_pred), max(y_pred)], color='red')
plt.show()
```

## 5. Interpreting the Mode

```python
print(f"Intercept (w₀): {model.intercept_}")
print("Coefficients (w₁, w₂, w₃):", model.coef_)

mse = mean_squared_error(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
```
