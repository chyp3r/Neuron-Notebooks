## 1. Supervised Learning

Supervised learning is a type of machine learning where the model is trained on labeled data, meaning the data includes both the input features (X) and the output labels (y). The goal is to learn a mapping from inputs to outputs so the model can make predictions on new, unseen data.

### Types of Supervised Learning

There are two main types of supervised learning:

- **Regression:** The model predicts a continuous value.
    - Example: Predicting house prices based on features like size, location, and number of rooms.
    - Algorithms: Linear Regression, Support Vector Regression (SVR), Decision Tree Regression.

- **Classification:** The model predicts a discrete label or category.
    - Example: Predicting whether an email is spam or not based on its content.
    - Algorithms: Logistic Regression, Random Forest, Support Vector Machines (SVM), k-Nearest Neighbors (k-NN).

### Example of Supervised Learning: Linear Regression

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.DataFrame({
    'feature1': np.random.rand(100),
    'feature2': np.random.rand(100),
    'target': np.random.rand(100) * 10 + 5  # Creating a continuous target variable
})
```

```python
X = data.drop(columns=["target"])
y = data["target"]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
print("Predictions:", y_pred[:5])
print("Real Values:", np.array(y_test[:5]))
```

## 2. Unsupervised Learning
In unsupervised learning, the model is trained on unlabeled data, meaning that the data only includes the input features (X) without the corresponding output labels (y). The model tries to discover patterns or structure in the data on its own.

### Types of Unsupervised Learning
There are two main types of unsupervised learning:

- Clustering: The model groups data into distinct clusters based on similarity.
	- Example: Grouping customers based on purchasing behavior.
	- Algorithms: K-Means Clustering, Hierarchical Clustering, DBSCAN.
- Dimensionality Reduction: The model reduces the number of features while preserving as much information as possible.
	- Example: Reducing a dataset with hundreds of features to a smaller set while retaining essential information.
	- Algorithms: Principal Component Analysis (PCA), t-SNE, Linear Discriminant Analysis (LDA).

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data_unsupervised = pd.DataFrame({
    'feature1': np.random.rand(100),
    'feature2': np.random.rand(100)
})

kmeans = KMeans(random_state =42)
kmeans.fit(data_unsupervised)

data_unsupervised['cluster'] = kmeans.labels_

plt.scatter(x=data_unsupervised["feature1"],y=data_unsupervised["feature2"],c=data_unsupervised['cluster'], cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## 3. Reinforcement Learning
Reinforcement learning is a type of machine learning where an agent interacts with an environment and learns to maximize a reward. The agent takes actions in an environment and receives feedback in the form of rewards or penalties. Over time, it learns the best strategy (called a policy) to maximize the total reward.

Key Concepts in Reinforcement Learning:
- Agent: The learner or decision-maker.
- Environment: The world in which the agent operates.
- Actions: The choices the agent can make.
- Rewards: Feedback from the environment after taking an action.
- Policy: The strategy the agent learns to maximize rewards.

Note: Reinforcement learning is more complex than supervised or unsupervised learning, and implementing it often requires simulation environments like OpenAI Gym.

Example libraries for reinforcement learning:
- `OpenAI Gym` for creating simulation environments.
- `Stable-Baselines3` for reinforcement learning algorithms.
