## Introduction to Matplotlib

`Matplotlib` is a powerful plotting library used to create a wide variety of charts and graphs.

```python
import matplotlib.pyplot as plt  # Importing the pyplot module from Matplotlib
import numpy as np  # Importing numpy for generating sample data

````python
import matplotlib.pyplot as plt
import numpy as np
````

## Basic Plotting

````python
x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y)
plt.title("Simple Sine Wave")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
````

## Customizing Plots

````python
plt.plot(x,y,color = "red",linestyle = "--",linewidth = 2,marker="o")
plt.title("Custom Sine Wave")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
````

##  Multiple Plots

````python
y2= np.cos(x)

plt.plot(x,y,label="Sine")
plt.plot(x,y2,label="Cosine")
plt.title("Sine and Cosine")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
````

## Bar Plots

````python
categories = ["A","B","C","D"]
values = [3,5,7,2]

plt.bar(categories,values,color="skyblue")
plt.title("Simple Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
````

## Scatter Plots

Scatter plots are used to visualize the relationship between two numerical variables.

````python
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

plt.scatter(x_scatter,y_scatter,color="green",marker="x")
plt.title("Simple Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
````

## Histograms

Histograms show the distribution of numerical data.

````python
hist_data = np.random.randn(1000)

plt.hist(hist_data,bins = 30,color="red",edgecolor="black")
plt.title("Simple Histogram Plot")
plt.xlabel("Data")
plt.ylabel("Hz")
plt.show()
````

## Pie Charts

Pie charts are used to show proportions of categorical data.

````python
labels = ["Apples","Bananas","Cherries","Dates"]
sizes = [15,30,45,10]

plt.pie(sizes,labels=labels,autopct="%1.1f%%",startangle=90)
plt.title('Simple Pie Chart')
plt.show()
````

## Working with Subplots

````python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1) # 1 row, 2 columns, 1st plot
plt.plot(x,y)
plt.title("Sine")

plt.subplot(1,2,2) # 1 row, 2 columns, 2nd plot
plt.plot(x,y2)
plt.title("Cosine")

plt.show()
````

## Saving Figures

````python
plt.plot(x,y)
plt.title("Sine Wave")
plt.savefig("sine_wave.png")
````
