## 1. Introduction to Seaborn

`Seaborn` provides beautiful and easy-to-use statistical plots. It works well with `Pandas` DataFrames and makes it easy to visualize trends, relationships, and patterns.

```python

````python
import seaborn as sns  # Importing seaborn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

We'll be using the `tips` dataset that comes built into Seaborn.

````python
# Load a dataset
tips = sns.load_dataset("tips")
````

````python
tips.head()
````

## Basic Plotting with Seaborn

````python
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("Total Bill vs Tip")
plt.show()
````

## Categorical Plots

````python
sns.barplot(x="day",y="total_bill",data=tips)
plt.title("Average Total Bill by Day")
plt.show()
````

````python
sns.boxplot(x="day",y="total_bill",data=tips)
plt.title("Total Bill Distribution by Day")
plt.show()
````

````python
sns.violinplot(x="day",y="total_bill",data=tips)
plt.title("Total Bill Distribution (Violin Plot)")
plt.show()
````

## Distribution Plots

````python
sns.histplot(tips["total_bill"],bins=30,kde=True)
plt.title("Distribution of Total Bills")
plt.show()
````

````python
sns.kdeplot(tips["total_bill"],fill=True)
plt.title("KDE Plot of Total Bills")
plt.show()
````

## Matrix Plots

Matrix plots are useful for visualizing relationships between data points, particularly when exploring correlations.

````python
corr = tips.corr(numeric_only=True)

sns.heatmap(corr,annot=True,cmap="coolwarm",linewidths=.5)
plt.title("Correlation Heatmap")
plt.show()
````

## Pair Plots

Pair plots are used to explore relationships between multiple variables. This is particularly useful for high-dimensional datasets.

````python
sns.pairplot(tips)
plt.show()
````

## Regression Plots

Regression plots are used to visualize the relationship between two variables and fit a linear regression model.

````python
sns.regplot(x="total_bill",y="tip",data=tips)
plt.title("Linear Regression: Total Bill vs Tip")
plt.show()
````

````python
sns.lmplot(x="total_bill",y="tip",hue="sex",data = tips)
plt.title("Linear Regression by Sex")
plt.show()
````

## Style Customization

````python
sns.set_theme(style="whitegrid",palette="pastel")

sns.swarmplot(x='day', y='total_bill', data=tips)
plt.title('Swarm Plot with Custom Style')
plt.show()
````

## Facet Grids

Facet Grids allow you to visualize subsets of your data across multiple plots.

````python
g = sns.FacetGrid(tips,col="time",row="sex")
g.map(sns.histplot,"total_bill")
plt.show()
````

## Saving Figures

````python
plt.figure()
sns.boxplot(x="day",y="total_bill",data=tips)
plt.savefig("boxplot_total_bill_by_day.png")
plt.show()
````
